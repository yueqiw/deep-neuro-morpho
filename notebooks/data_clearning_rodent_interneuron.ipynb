# Pool data together

```python
all_data_list = [interneuron_cortex, interneuron_hippo, interneuron_retina]
all_data = pd.concat(all_data_list).sort_values(by='neuron_id').reset_index(drop=True)
all_data.shape
```

```python
use_data_list = [interneuron_cortex[interneuron_cortex.cell_type_2_curated.isin(['basket', 'Martinotti'])],
                interneuron_retina]
use_data = pd.concat(use_data_list).sort_values(by='neuron_id').reset_index(drop=True)
use_data.shape
```

```python
use_data['label1'] = 'None'
```

```python
use_data.loc[use_data.cell_type_2_curated=='basket', 'label1'] = 'cortex basket'
use_data[use_data.cell_type_2_curated=='basket'].shape
```

```python
use_data.loc[use_data.cell_type_2_curated=='Martinotti', 'label1'] = 'cortex Martinotti'
use_data[use_data.cell_type_2_curated=='Martinotti'].shape
```

```python
use_data.loc[use_data.cell_type_2_curated=='bipolar', 'label1'] = 'retina bipolar'
use_data[use_data.cell_type_2_curated=='bipolar'].shape
```

```python
use_data.loc[use_data.cell_type_2_curated=='amacrine', 'label1'] = 'retina amacrine'
use_data[use_data.cell_type_2_curated=='amacrine'].shape
```
