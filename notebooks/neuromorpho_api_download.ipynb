# Download images with bad links in json metadata

## by scrapping the image file from html webpage

```python
import urllib3
from bs4 import BeautifulSoup
import re
import urllib
```

```python
http = urllib3.PoolManager()
```

```python
download_success = dict()
prefix = 'http://neuromorpho.org/'
for cid in neurons_no_png:
    cid = str(cid)
    #if int(cid) % 100 == 0:
    #    print(cid)
    http_address = 'http://neuromorpho.org/neuron_info.jsp?neuron_id=%s' % cid
    html = http.request('GET', http_address)
    soup = BeautifulSoup(html.data)
    imgfile = soup.find_all('img', src=re.compile('^./images/imageFiles'))
    if len(imgfile) != 1:
        print('image link error: %d image links found' % len(imgfile))
    sublink = imgfile[0].get('src')
    sublink = sublink.lstrip('.')
    sublink = sublink.lstrip('/')
    url = prefix + sublink
    url = url.replace(' ', '%20')
    print(url)
    download_success[cid] = True
    try:
        urllib.request.urlretrieve(url, "../data/png/%s.png" % cid)
    except:
        print('%s: download error.' % cid)
        download_success[cid] = False
```

```python
morph.json()
```

```python
morpho_allrequest = dict()
morpho_request = dict()
for i in range(1,70465):
    url = "http://neuromorpho.org/api/morphometry/id/%d.json" % i
    morpho = requests.get(url)
    if morpho.status_code != 200:
        print('%i: bad request status' % i)
    if i % 100 == 0:
        print(i)
    try:
        morpho_allrequest[i] = morpho.json()
    except:
        morpho_allrequest[i] = None
    morpho_request[i] = morpho.status_code
```

```python
with open('../data/morpho_allrequest.json', 'w') as f:
    json.dump(morpho_allrequest, f)
```

```python
morpho_all = {i: morpho_allrequest[i] for i in morpho_allrequest if not morpho_allrequest[i] is None}
```

# literature info

```python
pmid_all = []
for i in [0,1,2,3]:
    pmid_request = requests.get("http://neuromorpho.org/api/literature/fields/PMID.json?page=%d" % i)
    pmid_chunk = pmid_request.json()
    pmid_chunk = pmid_chunk['fields']
    pmid_all.extend(pmid_chunk)
```
