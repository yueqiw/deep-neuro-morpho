```python
domain_remove = ['No Dendrites, No Soma, Axon', 'No Dendrites, Soma, Axon']
```

```python
mammals_3d = mammals[mammals['3D'] & mammals['angles']]
mammals_3d = mammals_3d[~mammals_3d['domain'].isin(domain_remove)]
len(mammals_3d)
```

## some retina cells have neurites or processes, but we can still keep them

```python
integrity_list = list(mammals_3d['physical_Integrity'].unique())
print(len(integrity_list))
integrity_list
```

```python
good_dendrites = ['Dendrites Moderate',
                 'Dendrites Complete',
                 'Dendrites Complete, Axon Incomplete',
                 'Dendrites & Axon Moderate',
                 'Dendrites Complete, Axon Moderate',
                 'Dendrites & Axon Complete',
                 'Dendrites Moderate, Axon Incomplete',
                 'Dendrites Moderate, Axon Complete',
                 'Processes Moderate',
                 'Neurites Moderate']
complete_dendrites = ['Dendrites Complete',
                 'Dendrites Complete, Axon Incomplete',
                 'Dendrites Complete, Axon Moderate',
                 'Dendrites & Axon Complete']
moderate_dendrites = ['Dendrites Moderate',
                 'Dendrites & Axon Moderate',
                 'Dendrites Moderate, Axon Incomplete',
                 'Dendrites Moderate, Axon Complete',
                 'Processes Moderate',
                 'Neurites Moderate']
bad_dendrites = ['Dendrites & Axon Incomplete',
                 'Dendrites Incomplete',
                 'Dendrites Incomplete, Axon Moderate',
                 'Axon Incomplete',
                 'Dendrites Incomplete, Axon Complete']
print(len(good_dendrites), len(bad_dendrites))
```

```python
mammals_3d_den = mammals_3d[mammals_3d['physical_Integrity'].isin(good_dendrites)]
mammals_3d_completeden = mammals_3d[mammals_3d['physical_Integrity'].isin(complete_dendrites)]
mammals_3d_moderateden = mammals_3d[mammals_3d['physical_Integrity'].isin(moderate_dendrites)]
mammals_3d_badden = mammals_3d[mammals_3d['physical_Integrity'].isin(bad_dendrites)]
print(len(mammals_3d_den), len(mammals_3d_completeden), len(mammals_3d_moderateden), len(mammals_3d_badden))
```

```python
mammals_3d_den_full = mammals_3d_den.drop(['neuron_name', 'surface', 'volume', 'diameter'], axis=1).merge(morpho_df, on=['neuron_id'])
mammals_3d_completeden_full = mammals_3d_completeden.drop(['neuron_name', 'surface', 'volume', 'diameter'], axis=1).merge(morpho_df, on=['neuron_id'])
mammals_3d_moderateden_full = mammals_3d_moderateden.drop(['neuron_name', 'surface', 'volume', 'diameter'], axis=1).merge(morpho_df, on=['neuron_id'])
```

```python
mammals_3d_den_full.to_csv('../data/mammals_3d_dendrites.csv', index=False)
mammals_3d_completeden_full.to_csv('../data/mammals_3d_dendrites_complete.csv', index=False)
mammals_3d_moderateden_full.to_csv('../data/mammals_3d_dendrites_moderate.csv', index=False)
```
